```python
def fen_decode(fen: str):
    board = [['' for _ in range(8)] for _ in range(8)]
    parts = fen.split(' ')
    board_fen = parts[0]
    en_passant = parts[3]

    rank_ranks = board_fen.split('/')
    for i, rank_fen in enumerate(rank_ranks):
        file_index = 0
        for char in rank_fen:
            if char.isdigit():  # Number of empty files
                file_index += int(char)
            else:  # Piece
                if char.islower():
                    board[i][file_index] = char
                    file_index += 1
                elif char.isupper():
                    board[i][file_index] = char
                    file_index += 1

    for i in range(8):
        for j in range(8):
            if board[i][j] == '':
                board[i][j] = '.'

    return board, en_passant
```

```python
def check_pawn(board: list, color: bool) -> list:
    pawns = []
    moves = []
    
    for i in range(len(board)): # Get all pawns into list
        for j in range(len(board[i])):
            if board[i][j] == 'P' and color == True: # White
                pawns.append([j, i])
            elif board[i][j] == 'p' and color == False: # Black
                pawns.append([j, i])
    
    for pawn in pawns:
        if color:
            if board[pawn[1] - 1][pawn[0]] == '.':
                moves.append(f"P{chr(ord('a') + pawn[0])}{8 - pawn[1]}-{chr(ord('a') + pawn[0])}{8 - pawn[1] + 1}")
            if pawn[1] == 6 and board[pawn[1] - 2][pawn[0]] == '.' and board[pawn[1] - 1][pawn[0]] == '.':
                moves.append(f"P{chr(ord('a') + pawn[0])}{8 - pawn[1]}-{chr(ord('a') + pawn[0])}{8 - pawn[1] + 2}")
            if pawn[0] != 0:
                if board[pawn[1] - 1][pawn[0] - 1] != '.' and board[pawn[1] - 1][pawn[0] - 1] != 'k' and board[pawn[1] - 1][pawn[0] - 1].islower():
                    moves.append(f"P{chr(ord('a') + pawn[0])}{8 - pawn[1]}x{chr(ord('a') + pawn[0] - 1)}{8 - pawn[1] + 1}")
            if pawn[0] != 7:
                if board[pawn[1] - 1][pawn[0] + 1] != '.' and board[pawn[1] - 1][pawn[0] + 1] != 'k' and board[pawn[1] - 1][pawn[0] + 1].islower():
                    moves.append(f"P{chr(ord('a') + pawn[0])}{8 - pawn[1]}x{chr(ord('a') + pawn[0] + 1)}{8 - pawn[1] + 1}")
        else:
            if board[pawn[1] + 1][pawn[0]] == '.':
                moves.append(f"P{chr(ord('a') + pawn[0])}{8 - pawn[1]}-{chr(ord('a') + pawn[0])}{8 - pawn[1] - 1}")
            if pawn[1] == 1 and board[pawn[1] + 2][pawn[0]] == '.' and board[pawn[1] + 1][pawn[0]] == '.':
                moves.append(f"P{chr(ord('a') + pawn[0])}{8 - pawn[1]}-{chr(ord('a') + pawn[0])}{8 - pawn[1] - 2}")
            if pawn[0] != 0:
                if board[pawn[1] + 1][pawn[0] - 1] != '.' and board[pawn[1] + 1][pawn[0] - 1] != 'K' and board[pawn[1] + 1][pawn[0] - 1].isupper():
                    moves.append(f"P{chr(ord('a') + pawn[0])}{8 - pawn[1]}x{chr(ord('a') + pawn[0] - 1)}{8 - pawn[1] - 1}")
            if pawn[0] != 7:
                if board[pawn[1] + 1][pawn[0] + 1] != '.' and board[pawn[1] + 1][pawn[0] + 1] != 'K' and board[pawn[1] + 1][pawn[0] + 1].isupper():
                    moves.append(f"P{chr(ord('a') + pawn[0])}{8 - pawn[1]}x{chr(ord('a') + pawn[0] + 1)}{8 - pawn[1] - 1}")
    
    return moves
```

```python
def check_en_passant(board: list, color: bool, en_passant: str) -> str:
    pawns = []
    moves = []
    
    for i in range(len(board)): # Get all queens into list
        for j in range(len(board[i])):
            if board[i][j] == 'P' and color == True: # White
                pawns.append([j, i])
            elif board[i][j] == 'p' and color == False: # Black
                pawns.append([j, i])
    
    if en_passant != '-':
        if en_passant[0] == 'a':
            if color and board[3][1] == 'P':
                moves.append("Pb5xa6")
            elif color == False and board[4][1] == 'p':
                moves.append("Pb4xa3")
        elif en_passant[0] == 'h':
            if color and board[3][6] == 'P':
                moves.append("Pg5xh6")
            elif color == False and board[4][6] == 'p':
                moves.append("Pg4xah3")
        else:
            if color and board[3][ord(en_passant[0]) - ord('a') + 1] == 'P':
                moves.append(f"P{chr(ord(en_passant[0]) + 1)}5x{en_passant}")
            if color and board[3][ord(en_passant[0]) - ord('a') - 1] == 'P':
                moves.append(f"P{chr(ord(en_passant[0]) - 1)}5x{en_passant}")
            if color == False and board[4][ord(en_passant[0]) - ord('a') + 1] == 'p':
                moves.append(f"P{chr(ord(en_passant[0]) + 1)}4x{en_passant}")
            if color == False and board[4][ord(en_passant[0]) - ord('a') - 1] == 'p':
                moves.append(f"P{chr(ord(en_passant[0]) - 1)}4x{en_passant}")

    return moves
```
