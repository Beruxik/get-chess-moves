```python
def fen_decode(fen: str):
    board = [['' for _ in range(8)] for _ in range(8)]
    parts = fen.split(' ')
    board_fen = parts[0]
    en_passant = parts[3]

    rank_ranks = board_fen.split('/')
    for i, rank_fen in enumerate(rank_ranks):
        file_index = 0
        for char in rank_fen:
            if char.isdigit():  # Number of empty files
                file_index += int(char)
            else:  # Piece
                if char.islower():
                    board[i][file_index] = char
                    file_index += 1
                elif char.isupper():
                    board[i][file_index] = char
                    file_index += 1

    for i in range(8):
        for j in range(8):
            if board[i][j] == '':
                board[i][j] = '.'

    return board, en_passant
```

```python
def check_king(board: list, color: bool) -> None:
    kings = []
    for i in range(len(board)): # Get all kings into list
        for j in range(len(board[i])):
            if board[i][j] == 'K' and color == True: # White
                kings.append([j, i])
            elif board[i][j] == 'k' and color == False: # Black
                kings.append([j, i])

    for king in kings:
        if king[0] + 1 <= 7: # Check moves to the right
            if board[king[1]][king[0] + 1] == '.': # Empty square
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}-{chr(ord('a') + king[0] + 1)}{8 - king[1]}")
            elif (board[king[1]][king[0] + 1].islower() and color == True) or (board[king[1]][king[0] + 1].isupper() and color == False): # Takes
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}x{chr(ord('a') + king[0] + 1)}{8 - king[1]}")
            
        if king[0] - 1 >= 0: # Check moves to the left
            if board[king[1]][king[0] - 1] == '.': # Empty square
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}-{chr(ord('a') + king[0] - 1)}{8 - king[1]}")
            elif (board[king[1]][king[0] - 1].islower() and color == True) or (board[king[1]][king[0] - 1].isupper() and color == False): # Takes
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}x{chr(ord('a') + king[0] - 1)}{8 - king[1]}")
        
        if king[1] + 1 <= 7: # Check moves to the bottom
            if board[king[1] + 1][king[0]] == '.': # Empty square
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}-{chr(ord('a') + king[0])}{8 - king[1] - 1}")
            elif (board[king[1] + 1][king[0]].islower() and color == True) or (board[king[1] + 1][king[0]].isupper() and color == False): # Takes
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}x{chr(ord('a') + king[0])}{8 - king[1] - 1}")
        
        if king[1] - 1 >= 0: # Check moves to the top
            if board[king[1] - 1][king[0]] == '.': # Empty square
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}-{chr(ord('a') + king[0])}{8 - king[1] + 1}")
            elif (board[king[1] - 1][king[0]].islower() and color == True) or (board[king[1] - 1][king[0]].isupper() and color == False): # Takes
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}x{chr(ord('a') + king[0])}{8 - king[1] + 1}")
        
        if king[0] + 1 <= 7 and king[1] + 1 <= 7: # Check moves to the bottom right
            if board[king[1] + 1][king[0] + 1] == '.': # Empty square
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}-{chr(ord('a') + king[0] + 1)}{8 - king[1] - 1}")
            elif (board[king[1] + 1][king[0] + 1].islower() and color == True) or (board[king[1] + 1][king[0] + 1].isupper() and color == False): # Takes
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}x{chr(ord('a') + king[0] + 1)}{8 - king[1] - 1}")

        if king[0] + 1 <= 7 and king[1] - 1 >= 0: # Check moves to the top right
            if board[king[1] - 1][king[0] + 1] == '.': # Empty square
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}-{chr(ord('a') + king[0] + 1)}{8 - king[1] + 1}")
            elif (board[king[1] - 1][king[0] + 1].islower() and color == True) or (board[king[1] - 1][king[0] + 1].isupper() and color == False): # Takes
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}x{chr(ord('a') + king[0] + 1)}{8 - king[1] + 1}")

        if king[0] - 1 >= 0 and king[1] + 1 <= 7: # Check moves to the bottom left
            if board[king[1] + 1][king[0] - 1] == '.': # Empty square
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}-{chr(ord('a') + king[0] - 1)}{8 - king[1] - 1}")
            elif (board[king[1] + 1][king[0] - 1].islower() and color == True) or (board[king[1] + 1][king[0] - 1].isupper() and color == False): # Takes
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}x{chr(ord('a') + king[0] - 1)}{8 - king[1] - 1}")
        
        if king[0] - 1 >= 0 and king[1] - 1: # Check moves to the top left
            if board[king[1] - 1][king[0] - 1] == '.': # Empty square
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}-{chr(ord('a') + king[0] - 1)}{8 - king[1] + 1}")
            elif (board[king[1] - 1][king[0] - 1].islower() and color == True) or (board[king[1] - 1][king[0] - 1].isupper() and color == False): # Takes
                print(f"K{chr(ord('a') + king[0])}{8 - king[1]}x{chr(ord('a') + king[0] - 1)}{8 - king[1] + 1}")
```
